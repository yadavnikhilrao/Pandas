```python
# how to check duplicate rows by using duplicated() method

df = pd.read_csv("emp.csv")
df
```

```python
# keeps --> first -->  mark duplicates as true except for frst occurnce

# by default keep takes first

df["ename"].duplicated()
```

```python
# keeps--> last    --> mark duplicates as true except for the last occurnce

df["ename"].duplicated(keep = "last")
```

```python
# keeps--> last    --> mark all duplicates as true 

df["ename"].duplicated(keep = False)
```

```python
# select all rows without ename duplicates
# if ename contains same name consider only frst occurence

df = pd.read_csv("emp.csv")

c = ~df["ename"].duplicated()
df[c]
```

```python
#  ~  --> convert true into false , false into true
```

```python
# select all rows where ename is not repeated
# i.e -> select only unique enames


df = pd.read_csv("emp.csv")
df["ename"].unique()
```

```python
# how to drop duplicate rows

# drop_duplicates()

# by default it drop duplicate rows if all col values are duplicated
```

```python
df = pd.read_csv("emp1.csv")
df
```

```python
df.drop_duplicates()
```

```python
# for specified columns
# use subset parameter
# to dlete record where ename is duplicated

df.drop_duplicates(subset = ["ename"])
```

```python
df
```

```python
# to drop all records where ename is duplicated and won't consider
# frst occurnce

df.drop_duplicates(subset = ["ename"] , keep = False)
```
