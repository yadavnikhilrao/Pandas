```python
# how to rename index labels and col name in a data frame
# index label - axis 0
# col names  - axis 1
```

```python
# by using rename
# renaming of index labels
# by using mapper and axis parameter (axis 1 - index , axis 2 - columns)
```

```python
# by using index parameter

df = pd.read_csv("emp.csv" , index_col = "ename")
df
```

```python
# chnge index name tina to ram

df.rename(mapper = {"riya":"ram"})
```

```python
# chnge index name arun to anil and renu to pooja

df.rename(mapper = {"arun":"anil" , "riya":"pooja"})
#or
df.rename(mapper = {"arun":"ankit" , "riya":"pooja"},axis = 0 )
# or
df.rename(mapper = {"arun":"ankit" , "riya":"pooja"},axis = "index" )
# or
df.rename(mapper = {"arun":"ankit" , "riya":"pooja"},axis = "rows" )
```

```python
# 2. by using index parameter

df = pd.read_csv("emp.csv" , index_col = "ename")
df
```

```python
df.rename(index = {"arun":"riya","riya":"pihu"})
```

```python
# renaming colmun names

# by using mapper
# by using columns parameter


import pandas as pd

df = pd.read_csv("emp.csv" , index_col = "ename")
df.head()
```

```python
# by using mapper
# here we need to pass axis = 1 for columns

df.rename(mapper = {"esal":"sal" , "eadrs":"addrs"} , axis = 1)
```

```python
df.rename(mapper = {"esal":"sal" , "eadrs":"addrs"})
#or
df.rename(mapper = {"esal":"sal" , "eadrs":"addrs"},axis = 1 )
# or
df.rename(mapper = {"esal":"sal" , "eadrs":"addrs"},axis = "columns" )
```

```python
# by using columns parameter

df.rename(columns = {"esal":"sal" , "eadrs":"addr"})
```

```python
# shortcut to rename columns
# we can use column attribute of data frame


df.columns = ["ENO","SALARY","CITY"]   # no of col should match with oringl no of cols
```

###  delete rows and columns from a dataframe

```python
# we have to use drop method
# delete rows

import pandas as pd
df = pd.read_csv("emp.csv" , index_col = "ename")
df.head()
```

```python
df.drop(labels="arun")
# or
df.drop(labels = "arun" , axis = 0)
# or
df.drop(labels = "arun" , axis = "index")
```

```python
# or
df.drop(index = ["arun" , "nikhil" ,"prakriti"])
```

```python
# delete columns
# we can use either label and axis or column parameter 

import pandas as pd
df = pd.read_csv("emp.csv" , index_col = "ename")
df.head()
```

```python
df.drop(labels = ["esal"] , axis = 1)
# or
df.drop(labels = ["esal"] , axis = "columns")
# or
df.drop(columns = ["eadrs" , "esal"])     # execute last one
```

```python
import pandas as pd
df = pd.read_csv("emp.csv" , index_col = "ename")
df.head()
```

```python
# directly we can dlte rows

df.drop("arun")
df.drop("nikhil",axis = 0)  # for rows
df.drop(["arun" , "nikhil"] , axis = "rows")
df.drop(["arun" , "nikhil"] , axis = "index")
```

```python
# for columns

df.drop("eno",axis = 1)  
df.drop(["esal" , "eno"] , axis = "columns")
```
