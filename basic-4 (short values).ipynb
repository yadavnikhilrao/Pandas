```python
# how to rank series of values by using rank() method
# compute numerical data ranks (1 to n)
```

```python
df = pd.read_csv("emp.csv")
df
```

```python
df["esal"].rank()   # least salary rank 1 (ascending order)
```

```python
df["esal"].rank(ascending = False)  # highest salary rank 1 (descending order)
```

```python
# defualt dtype of rank is float
#  if equal values are there 
#  then it will take average value of ranks of those sal

# ex - values of rank1 rank 2 are same 
# rank 1 + rank 2 = rank3
# avg = rank 1.5
# sp rank 1.5 is assignd to both of them
```
