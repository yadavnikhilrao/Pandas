```python
# how to drop rows with missing / null values

# dropna() method
```

```python
import pandas as pd

df = pd.read_csv("emp.csv")
df
```

```python
# df.dropna() - by default it  removes rows from the dataframe where atleast one missing value
```

```python
df.dropna()
```

```python
# how parameter --> 
# any - if any na values are present , drop that row 
# all - if all values are na , drop that row 
```

```python
df.dropna(how = "any")

#df.dropna(how = "all")
```

```python
# subset parameter
# to delte rows if par col has missing values

# only for dltng rows
```

```python
import pandas as pd

df = pd.read_csv("emp.csv")
df
```

```python
df.dropna(subset = ["esal" , "eadrs"])
```

```python
df.dropna(subset = ["esal"])
```

### fill null values

```python
# how to fill missing / null values with our rqrd value in dataframe
```

```python
# fillna()
```

```python
df = pd.read_csv("fill.csv")
df
```

```python
# but this is meaning less as adrs col should not contain 0 as it takes string value
```

```python
# replace ename with nikhil
# replace eadrs with delhi
# replace esal with 10000
```

```python
df
```
