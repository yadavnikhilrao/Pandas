```python
# while performing joins , if we want to consider only
# common  columns , then we have to use join paramter

# join - inner , outer
# by default outer is considered
```

```python
pd.concat([df1,df2] , join = 'outer')
```

```python
pd.concat([df1,df2] , join = "inner")
```
